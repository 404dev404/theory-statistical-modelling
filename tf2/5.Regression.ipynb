```python
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers

print(tf.__version__)

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
dataset_path = keras.utils.get_file("auto-mpg.data", "http://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data")
dataset_path
```

```python
column_names = ['MPG','Cylinders','Displacement','Horsepower','Weight',
                'Acceleration', 'Model Year', 'Origin'] 

raw_dataset = pd.read_csv(dataset_path, names = column_names,
                         na_values = "?", comment = "\t",
                         sep = " ", skipinitialspace = True)
```

```python
dataset = raw_dataset.copy()
dataset.tail()
```

```python
dataset = dataset.dropna()
```

```python
origin = dataset.pop("Origin")
```

```python
dataset["USA"] = (origin == 1)*1.0
dataset["Europe"] = (origin == 2)*1.0
dataset["Japan"] = (origin == 3)*1.0

dataset.tail()
```

```python
train_dataset = dataset.sample(frac=0.8, random_state = 0)
test_dataset = dataset.drop(train_dataset.index)
```

```python
sns.pairplot(train_dataset[["MPG", "Cylinders", "Displacement", "Weight"]], diag_kind = "kde")
```

```python
train_stats = train_dataset.describe()
train_stats.pop("MPG")

train_stats = train_stats.transpose()
train_stats
```

```python
train_labels = train_dataset.pop('MPG')
test_labels = test_dataset.pop('MPG')
```

```python
train_dataset.head(5)
```

```python
def norm(x):
    return (x - train_stats['mean']) / train_stats['std']

normed_train_data = norm(train_dataset)
normed_test_data = norm(test_dataset)
```

```python

normed_train_data.head(5)
```

```python
import numpy as np
```

```python
max(normed_train_data["Horsepower"])
```

```python
# denormalize
-0.784052 * 38.096214 + 104.869427 	
```

```python
def build_model():
    model = keras.Sequential([
        layers.Dense(64, activation = "relu", input_shape = [len(train_dataset.keys())]),
        layers.Dense(64, activation = "relu"),
        layers.Dense(1)
    ])
    
    optimizer = tf.keras.optimizers.RMSprop(0.001)

    model.compile(loss = "mse",
                 optimizer = optimizer,
                 metrics = ['mae', 'mse'])
    return model
```

```python
model = build_model()
```

```python
model.summary()
```

```python
example_batch = normed_train_data[:10]
example_result = model.predict(example_batch)
```

```python
# Display training progress by printing a single dot for each completed epoch
class PrintDot(keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs):
        if epoch % 100 == 0: print('')
        print('.', end='')

EPOCHS = 1000

history = model.fit(
  normed_train_data, train_labels,
  epochs=EPOCHS, validation_split = 0.2, verbose=0,
  callbacks=[PrintDot()])
```

```python
hist = pd.DataFrame(history.history)
hist['epoch'] = history.epoch
hist.tail()
```

```python
def plot_history(history):
    hist = pd.DataFrame(history.history)
    hist["epoch"] = history.epoch
    
    plt.figure(figsize = (14, 10))
    plt.xlabel("Epoch", color = "white")
    plt.ylabel("Mean Abs Error [MPG]", color = "white")
    plt.plot(hist['epoch'], hist['mae'],
            label = "Train Error")
    plt.plot(hist['epoch'], hist['val_mae'],
            label = "Val Error")
    plt.ylim([0, 5])
    plt.legend()
    plt.xticks(color = "white")
    plt.yticks(color = "white")
    
    plt.figure(figsize=(14, 10))
    plt.xlabel('Epoch', color = "white")
    plt.ylabel('Mean Square Error [$MPG^2$]', color = "white")
    plt.plot(hist['epoch'], hist['mse'],
           label='Train Error')
    plt.plot(hist['epoch'], hist['val_mse'],
           label = 'Val Error')
    plt.ylim([0, 20])
    plt.legend()
    plt.yticks(color = "white")
    plt.xticks(color = "white")
    plt.show()

    
```

```python
plot_history(history)
```

```python
model = build_model()
```

```python
early_stop = keras.callbacks.EarlyStopping(monitor = 'val_loss', patience = 10)

history = model.fit(normed_train_data, train_labels, epochs = EPOCHS,
                   validation_split = 0.2, verbose = 0, callbacks = [early_stop, PrintDot()])
```

```python
plot_history(history)
```

```python
test_predictions = model.predict(normed_test_data).flatten()

plt.figure(figsize=(14, 10))
plt.scatter(test_labels, test_predictions)
plt.xlabel('True Values [MPG]', color = "white")
plt.ylabel('Predictions [MPG]', color = "white")
plt.axis('equal')
plt.axis('square')
plt.xlim([0,plt.xlim()[1]])
plt.ylim([0,plt.ylim()[1]])
_ = plt.plot([-100, 100], [-100, 100])
plt.xticks(color = "white")
plt.yticks(color = "white")
```

```python
error = test_predictions - test_labels

plt.figure(figsize=(14, 10))
plt.hist(error, bins = 25)
plt.xlabel("Prediction Error [MPG]", color = "white")
plt.xticks(color = "white")
plt.yticks(color = "white")
plt.ylabel("Count", color = "white")
plt.show()
```
